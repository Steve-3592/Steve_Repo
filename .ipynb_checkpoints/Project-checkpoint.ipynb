 ---

__1. THE HISTOGRAMS__

The histograms shows the distribution of some selected feature variables of the red wine dataset.
Based on histograms of columns in this dataset, Fixed Acidity and Alcohol feature variables appear skewed to the right.
The direction of skew is shown by the length of the tails of the distribution. A long right tail means skewed to the right.
What does this mean in practical terms? It means that there are some outliers on the right side of the distribution
The effect is that these outliers, literally "extreme observations", will "drag" the mean/average up
(so that it will be less representative of the data as a whole).

__2. THE SCATTER PLOTS__

The scatter plots show the distribution of quality vs some selected feature variables of the red wine dataset(volatile acidity, residual sugar, pH and alcohol). It shows how the selected feature variables impact the quality of the wine. We can find the feature variable with higher possibility of impacting the wine quality.This means we are looking for a positive correlation and high value of correlation coefficient.Base on the scatter plots, alcohol has a positve correlation and highest among all the positive correlation scatterplots. The higher the alcohol content, the higher the quality.
This implies that alcohol has the  higher possibility of impacting the wine quality.

__3. THE HEATMAP__


The heatmap shows the correlation coefficients for the feature variables in the wine dataset.This is how the heatmap works: Calculates the correlation matrix for the dataset, Sets the figure size for the heatmap, Plots the heatmap with annotations (actual correlation coefficients) on each cell, Rotates the x-axis labels for better visibility of the heatmap.
This visualization is helpful for understanding the pairwise correlations between different feature variables in the dataset. The color intensity and the values in each cell indicate the strength and direction of the correation.
Referring to the bottom row of Quality, the feature variable with the most positive impact on Quality is Alcohol at 0.44 coefficient.
The other features such as Volatile Acidity, Residual Sugar and pH have negative or less than 0.1 correlation.l

__4. THE PAIRPLOT___

The pairplot showcase the relationships between selected features('alcohol', 'pH', 'residual sugar', 'citric acid') with different wine quality levels distinguished by color..
The title is set to 'Pairplot of Selected Features by Wine Quality'. The plt.show() command displays the plot.
This pairplot is useful for visually inspecting how the selected features vary in relation to each other across different wine quality levels. 
The diagonal represents the distribution of each individual variable, while the scatter plots show the relationships between pairs of variables.
